```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chess  = cv2.imread("Computer-Vision-with-Python/DATA/flat_chessboard.png")
chess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
plt.imshow(chess)
```

```python
gray_flat = cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat,cmap="gray")
```

```python
real_chess = cv2.imread("Computer-Vision-with-Python/DATA/real_chessboard.jpg")
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real,cmap="gray")
```

```python
gray = np.float32(gray_flat)
```

```python
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
chess[dst>0.01*dst.max()] = [0,0,255]
```

```python
plt.imshow(chess)
```

```python
gray = np.float32(gray_real)
```

```python
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()] = [0,0,255]
```

```python
plt.imshow(real_chess)
```

```python
room = cv2.imread("Computer-Vision-with-Python/DATA/bedroom-interior-designing.jpg")
room = cv2.cvtColor(room,cv2.COLOR_BGR2RGB)
plt.imshow(room)
```

```python
gray_room = cv2.cvtColor(room,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_room,cmap="gray")
```

```python
gray = np.float32(gray_room)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst,None)
room[dst>0.01*dst.max()] = [255,0,0]
plt.imshow(room)
```
