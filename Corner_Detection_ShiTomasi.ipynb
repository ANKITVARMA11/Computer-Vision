```python
import cv2
import numpy as nm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread("Computer-Vision-with-Python/DATA/flat_chessboard.png")
real_chess = cv2.imread("Computer-Vision-with-Python/DATA/real_chessboard.jpg")


flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)


flat_gray = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
real_gray = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

# The ShiTomasi way of finding corners is detailed in the paper "Good Features to Track". Interestingly, the method in OpenCV that uses this way to find corners is also named as cv2.goodFeaturesToTrack

```python
corners = cv2.goodFeaturesToTrack(flat_gray,10,0.01,100)

# image: MatLike,
# maxCorners: int, MAXIMUM NUMBER OF CORNERS
# qualityLevel: float, THIS VALUE IS MULTIPLIED BY THE BEST CORNER QUALITY VALUE (BASICALLY LEAST EIGEN VALUE)
# minDistance: float, 
# corners: MatLike | None = ...,
# mask: MatLike | None = ...,
# blockSize: int = ...,
# useHarri
```

```python
corners = nm.int_(corners)
corners
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess,(x,y),2,(0,0,255),thickness=2) #RGB
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(real_gray,80,0.01,10)
corners = nm.int_(corners)
```

```python
corners
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess, (x,y),2,(0,0,255), thickness = 2)
plt.imshow(real_chess)
```

```python
room = cv2.imread("Computer-Vision-with-Python/DATA/bedroom-interior-designing.jpg")
room = cv2.cvtColor(room,cv2.COLOR_BGR2RGB)

gray_room = cv2.cvtColor(room,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray_room,0,0.1,1)
corners = nm.int_(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(room,(x,y),2,(255,0,0),10)
plt.imshow(room)
```
