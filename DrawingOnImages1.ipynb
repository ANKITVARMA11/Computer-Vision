```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank = np.zeros((512,512,3), dtype = np.int16)
```

```python
type(blank)
```

```python
blank.shape
```

```python
plt.imshow(blank)
```

```python
cv2.rectangle(blank, pt1=(3520,0),pt2=(150,150),color =(0,255,0),thickness=10)
```

```python
plt.imshow(blank)
```

```python
cv2.rectangle(blank, pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=2)
plt.imshow(blank)
```

```python
cv2.circle(img = blank, center=(250,250), radius = 50, color = (255,255,0),thickness=5)
plt.imshow(blank)
```

```python
cv2.line(blank, pt1= (0,0), pt2=(512,512),color=(232,134,254),thickness=5)
plt.imshow(blank)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank, text= "Hello, This is a text sample",org=(10,500), fontFace=font, fontScale=2,color=(255,0,200),thickness = 3, lineType=cv2.LINE_AA)
plt.imshow(blank)
```

```python
blank = np.zeros(shape=(512,512,3),dtype = np.int32)
```

```python
plt.imshow(blank)
```

```python
vertices = np.array([[100,400],[200,200],[400,400],[300,200]],dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
vertices
```

```python
pts
```

```python
cv2.polylines(blank,[pts], isClosed = True, color=(255,0,0), thickness= 3)
plt.imshow(blank)
```
