```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = "gray")
```

```python
img = cv2.imread("Computer-Vision-with-Python\DATA\sudoku.jpg",0)
```

```python
show_pic(img)
```

Sobel is a gradient type. When xgradient is calculated using sobel vertical lines are highly visible and vice-versa

```python
sobelx  = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 5)
```

```python
show_pic(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
show_pic(sobely)
```

Laplacian is yet another gradient detection type

CAN also be considered as EDGE DETECTION TYPE

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
show_pic(laplacian)
```

```python
blended = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
show_pic(blended)
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT, kernel)
show_pic(gradient)
```
