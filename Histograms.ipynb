```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = "gray")
```

```python
dark_horse = cv2.imread('Computer-Vision-with-Python\DATA\horse.jpg')  #Original BGR for OpenCV
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) # Converted to RGB for show

rainbow = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
show_pic(show_bricks)
```

```python
# BGR for OpenCV
hist_values = cv2.calcHist([blue_bricks],channels = [0],mask=None, histSize=[256],ranges = [0,256])
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
# BGR for OpenCV
hist_values = cv2.calcHist([dark_horse],channels = [0],mask=None, histSize=[256],ranges = [0,256])
```

```python
plt.plot(hist_values)
```

```python
img = dark_horse
```

```python
color = ("b","g","r")
for i, col in enumerate(color):
    histr = cv2.calcHist([img],channels = [i],mask=None, histSize=[256],ranges = [0,256])
    plt.plot(histr,col)
    plt.xlim([0,50])
    plt.ylim([0,500000])

plt.title("Histogram For dark horse")
```

```python
rainbow = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
mask[300:400,100:400] = 255
```

```python
show_pic(mask)
```

```python
masked_img = cv2.bitwise_and(img,img,mask = mask)
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow,show_rainbow,mask = mask)
```

```python
show_pic(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow],channels=[2], mask=mask,histSize=[256],ranges=[0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow],channels=[2], mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title("Red Histogram for Masked Rainbow")
```

```python
plt.plot(hist_values_red)
plt.title("Histogram for non - Masked Rainbow")
```

# EQUALIZING THE HISTOGRAM

```python
gorilla = cv2.imread("Computer-Vision-with-Python\DATA\gorilla.jpg",0)
```

```python
gorilla.shape
```

```python
show_pic(gorilla)
```

```python
hist_values = cv2.calcHist([gorilla], channels = [0], mask = None, histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
show_pic(eq_gorilla)
```

```python
hist_values = cv2.calcHist([eq_gorilla], channels = [0], mask = None, histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
color_gorilla = cv2.imread("Computer-Vision-with-Python\DATA\gorilla.jpg")
```

```python
show_pic(color_gorilla)
```

```python
hsv = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2HSV)
```
