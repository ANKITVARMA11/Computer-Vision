```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/rainbow.jpg")
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap = "gray")
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap = "gray")
```

```python
img = cv2.imread("Computer-Vision-with-Python\DATA\crossword.jpg",0)
```

```python
plt.imshow(img,cmap="gray")
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = "gray")
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 150,255, cv2.THRESH_BINARY)
show_pic(th1)
```

## ADAPTIVE THRESHOLDING

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,blockSize=11, C= 8)
show_pic(th2)
```

We can try blending both of these images to get some proper visual

```python
blended = cv2.addWeighted(th1,0.6,th2,0.2,0)
show_pic(blended) 
```
