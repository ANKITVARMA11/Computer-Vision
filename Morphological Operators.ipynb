```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,text = "ANKIT", org = (50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness = 15 )
    return blank_img
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = "gray")
```

```python
img = load_img()
show_pic(img)
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
```

Erode function applies the erosion by using a same kernel and passing it over the image iterating multiple times.

```python
result = cv2.erode(img,kernel,iterations = 1)
show_pic(result)
```

```python
result = cv2.erode(img,kernel,iterations = 3)
show_pic(result)
```

Opening is nothing but Erosion followed by Dialation. Useful to reduce Background noise. It's a morphological operator

```python
img = load_img()
```

```python
white_noise = np.random.randint(low = 0, high = 2, size = (600,600))
```

```python
white_noise #Just adding some random noise
```

```python
show_pic(white_noise)
```

```python
img.max()
```

```python
white_noise = white_noise * 255
```

```python
show_pic(white_noise)
```

```python
noise_img = white_noise + img
```

```python
show_pic(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN, kernel)
```

```python
show_pic(opening)
```

Erosion : Getting the rid between background and foreground edge
Dialation : Adding extra content(bubbles) to the edges
Morph Gradient : It consideres the difference between above two and hence can be used for Edge Detection

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
show_pic(gradient)
```
