cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE), cv2.drawContours(external_contours, contours,i,255,1) are the 2 functions used to find and draw contours in the image.

NOTE :
1. "findContours" returns 2 values (contours, heirarchy) for cv2,cv4 function while for cv3, returns 3 values (image, contours, heirarchy)
2. Read the input image as grayimage or convert it using cvtColor or apply a threshold on the image and then send it to find contours in.
3. For "drawContours" function, i value can be changed based on the category of xontours required

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Computer-Vision-with-Python\DATA\internal_external.png",0)
```

```python
plt.imshow(img,cmap = "gray")
```

```python
contours, heirarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_contours = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    if heirarchy[0][i][3] == 4:
        cv2.drawContours(external_contours, contours,i,255,1)



plt.imshow(external_contours,cmap = "gray")
```
